```python
%reload_ext rubberize
from rubberize import config
import pint

ureg = pint.UnitRegistry()
```

```python
%%tap
import math

# **Input parameters**
b = 400 * ureg.mm                   # beam width
h = 600 * ureg.mm                   # beam depth
L_n = 5.6 * ureg.m                  # beam clear span
c_c = 40 * ureg.mm                  # clear concrete cover
d_st = 10 * ureg.mm                 # diameter of stirrups
d_b = 16 * ureg.mm                  # diameter of main bar
d = h - c_c - d_st - 1/2 * d_b      # beam effective depth
f_cp = 27.6 * ureg.MPa              # 28-day concrete compressive strength
f_y = 415 * ureg.MPa                # main bar yield strength
f_yt = 275 * ureg.MPa               # transverse bar yield strength
M_u = 200 * (ureg.kN * ureg.m)      # factored moment
V_u = 200 * ureg.kN                 # factored shear
phi = 0.90                          # strength reduction factor
```

```python
%%tap
# **Calculations: Flexure**
R_n = (M_u/(phi * b * d**2)).to(ureg.MPa)
m = f_y/(0.85 * f_cp)
rho = 1/m * (1 - math.sqrt(1 - (2*m*R_n)/f_y))
rho_min = max(1.4*ureg.MPa, 0.25*math.sqrt(f_cp/ureg.MPa)*ureg.MPa)/f_y
rho = max(rho, rho_min)
A_s = rho*b*d
A_b = math.pi/4 * d_b**2
n = math.ceil(A_s/A_b)
```

```python
%%tap
# **Calculations: Integrity reinforcement**
A_s = n * A_b
beta_1 = 0.85 if 17 * ureg.MPa <= f_cp <= 28 * ureg.MPa else 0.65 if f_cp > 56 * ureg.MPa else 0.85 - 0.05/(7 * ureg.MPa) * (f_cp - 28 * ureg.MPa)
c = (A_s * f_y)/(0.85 * f_cp * beta_1 * b)
f_s = (600 * ureg.MPa) * (d - c)/c
f_s > f_y # Tension steel yields
phi = 0.90 if f_s >= 1000 * ureg.MPa else 0.65 if f_s < f_y else 0.65 + 0.25 * (f_s - f_y)/(1000 * ureg.MPa - f_y) # Beam is tension-controlled
a = beta_1 * c
phiM_np = (phi * A_s * f_y * (d - a/2)).to(ureg.kN * ureg.m) # Positive factored moment
phiM_nn = phiM_np/2 # Negative factored moment
R_n = (phiM_nn/(phi * b * d**2)).to(ureg.MPa)
m = f_y/(0.85 * f_cp)
rho = 1/m * (1 - math.sqrt(1 - (2*m*R_n)/f_y))
rho_min = max(1.4*ureg.MPa, 0.25*math.sqrt(f_cp/ureg.MPa)*ureg.MPa)/f_y
rho = max(rho, rho_min)
A_s = rho*b*d
A_b = math.pi/4 * d_b**2
n = math.ceil(A_s/A_b)
```
